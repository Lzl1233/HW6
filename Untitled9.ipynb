```python
###Q1

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gammaln
from math import log, pi, sqrt


np.random.seed(123)
n = 100


mu_true = 0.0
sigma_true = 1.0
nu_true = 5

from scipy.stats import t as tdist
z_data = mu_true + sigma_true * tdist(df=nu_true).rvs(n)

μ0   = 0.0
τ0   = 10.0
σ0   = 1.0
max_nu = 100
def log_prior_mu(mu):
    return -0.5 * ((mu - μ0)/τ0)**2 - log(τ0 * sqrt(2*pi))

def log_prior_sigma(sigma):
    if sigma <= 0:
        return -np.inf
    return -0.5*(sigma**2)/(σ0**2)

def log_prior_nu(nu):
    if (nu < 1) or (nu > max_nu) or (int(nu) != nu):
        return -np.inf
    return -log(max_nu)


def log_lik(mu, sigma, nu, z):
    if sigma <= 0 or nu <= 0:
        return -np.inf
    c1 = gammaln((nu+1)/2) - gammaln(nu/2) - 0.5*log(nu*pi) - log(sigma)
    ll = 0.0
    for zi in z:
        t2 = (zi - mu)/sigma
        ll += c1 - ((nu+1)/2)*log(1 + (t2**2)/nu)
    return ll

def log_post(mu, sigma, nu, z):
    return log_lik(mu, sigma, nu, z) \
           + log_prior_mu(mu) \
           + log_prior_sigma(sigma) \
           + log_prior_nu(nu)

n_iter = 5000
burn_in = 1000

chain_mu    = np.zeros(n_iter)
chain_sigma = np.zeros(n_iter)
chain_nu    = np.zeros(n_iter)

current_mu    = 0.0
current_sigma = 1.0
current_nu    = 5

current_log_post = log_post(current_mu, current_sigma, current_nu, z_data)

proposal_sd_mu    = 0.5
proposal_sd_sigma = 0.3

accept_mu    = 0
accept_sigma = 0
accept_nu    = 0

for t in range(n_iter):
    proposal_mu = np.random.normal(current_mu, proposal_sd_mu)
    proposal_log_post = log_post(proposal_mu, current_sigma, current_nu, z_data)
    log_alpha = proposal_log_post - current_log_post
    if log_alpha >= 0 or np.log(np.random.rand()) < log_alpha:
        current_mu = proposal_mu
        current_log_post = proposal_log_post
        accept_mu += 1

    proposal_sigma = np.random.normal(current_sigma, proposal_sd_sigma)
    proposal_log_post = log_post(current_mu, proposal_sigma, current_nu, z_data)
    log_alpha = proposal_log_post - current_log_post
    if log_alpha >= 0 or np.log(np.random.rand()) < log_alpha:
        current_sigma = proposal_sigma
        current_log_post = proposal_log_post
        accept_sigma += 1

    step_nu = np.random.choice([-1, 0, 1])
    proposal_nu = current_nu + step_nu
    
    proposal_log_post = log_post(current_mu, current_sigma, proposal_nu, z_data)
    log_alpha = proposal_log_post - current_log_post
    if log_alpha >= 0 or np.log(np.random.rand()) < log_alpha:
        current_nu = proposal_nu
        current_log_post = proposal_log_post
        accept_nu += 1

    chain_mu[t]    = current_mu
    chain_sigma[t] = current_sigma
    chain_nu[t]    = current_nu

acc_rate_mu    = accept_mu    / n_iter
acc_rate_sigma = accept_sigma / n_iter
acc_rate_nu    = accept_nu    / n_iter
print(f"Acceptance rate for mu:    {acc_rate_mu:.3f}")
print(f"Acceptance rate for sigma: {acc_rate_sigma:.3f}")
print(f"Acceptance rate for nu:    {acc_rate_nu:.3f}")

chain_mu_burn    = chain_mu[burn_in:]
chain_sigma_burn = chain_sigma[burn_in:]
chain_nu_burn    = chain_nu[burn_in:]

mu_mean    = np.mean(chain_mu_burn)
sigma_mean = np.mean(chain_sigma_burn)
counts_nu  = np.bincount(chain_nu_burn.astype(int))
nu_mode    = np.argmax(counts_nu)

print(f"Posterior mean of mu = {mu_mean:.3f}")
print(f"Posterior mean of sigma = {sigma_mean:.3f}")
print(f"Posterior mode of nu = {nu_mode}")

fig, axs = plt.subplots(3,2, figsize=(10,8))

axs[0,0].plot(chain_mu_burn, 'b-', alpha=0.5)
axs[0,0].set_title("Trace of mu")
axs[1,0].plot(chain_sigma_burn, 'r-', alpha=0.5)
axs[1,0].set_title("Trace of sigma")
axs[2,0].plot(chain_nu_burn, 'g-', alpha=0.5)
axs[2,0].set_title("Trace of nu")

axs[0,1].hist(chain_mu_burn, bins=30, color='blue', alpha=0.7)
axs[0,1].set_title("Posterior of mu")
axs[1,1].hist(chain_sigma_burn, bins=30, color='red', alpha=0.7)
axs[1,1].set_title("Posterior of sigma")
axs[2,1].hist(chain_nu_burn, bins=range(1,max_nu+2), color='green', alpha=0.7)
axs[2,1].set_title("Posterior of nu")

plt.tight_layout()
plt.show()
```

```python
###Q2

import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from scipy.stats import t as tdist

def generate_t_data(n, mu, sigma, nu, seed=42):
    """
    Generate 'n' observations from a location-scale t distribution:

        z_i = mu + sigma * T_{nu},

    where T_{nu} denotes a standard t distribution with 'nu' degrees of freedom.
    """
    np.random.seed(seed)
    return mu + sigma * tdist(df=nu).rvs(size=n)

def run_pymc_inference(n_list, mu_true=0.0, sigma_true=1.0, nu_true=5):
    """
    For each sample size in 'n_list' (e.g., [20, 50, 100, 500]),
    generate data from a location-scale t distribution, 
    build a PyMC model, and perform MCMC inference. 
    Then return the collected traces for analysis.
    """
    results = {}
    
    for n in n_list:
        print(f"\n=== Posterior Inference for n = {n} ===")
        
        z_data = generate_t_data(n, mu_true, sigma_true, nu_true, seed=123 + n)

        with pm.Model() as model:
            mu = pm.Normal("mu", mu=0, sigma=10)

            sigma = pm.HalfNormal("sigma", sigma=1)

            nu = pm.DiscreteUniform("nu", lower=1, upper=101)

            obs = pm.StudentT("obs", nu=nu, mu=mu, sigma=sigma, observed=z_data)

            step_nu  = pm.Metropolis(vars=[nu])
            step_all = [step_nu]
            
            trace = pm.sample(draws=4000,
                              tune=2000,
                              step=step_all,
                              chains=2,
                              cores=2,
                              target_accept=0.9,
                              return_inferencedata=True,
                              progressbar=True)
        
        results[n] = trace
    
    return results

n_list = [20, 50, 100, 500]
all_traces = run_pymc_inference(n_list)

for n in n_list:
    trace_n = all_traces[n]
    
    nu_samples = trace_n.posterior["nu"].stack(draws=("chain","draw")).values
    
    print(f"\n=== Summary for n={n} ===")
    print(az.summary(trace_n, var_names=["mu","sigma","nu"], kind="stats"))
    
    plt.figure()
    counts = np.bincount(nu_samples.astype(int))
    plt.bar(range(len(counts)), counts, width=0.7, color="tab:blue", alpha=0.7)
    plt.title(f"Posterior of nu (n={n})")
    plt.xlabel("nu")
    plt.ylabel("Frequency in posterior sample")
    plt.xlim([0, 105])
    plt.show()
```

```python
###Q3

import numpy as np
from scipy.special import gammaln
import math
import matplotlib.pyplot as plt

# ---------------------------
# 1) Synthetic Data
# ---------------------------
np.random.seed(123)
n = 100
mu_true = 0.0
sigma_true = 1.0
nu_true = 5

# Generate t_{nu_true} via standard library:
from scipy.stats import t as tdist
z_data = mu_true + sigma_true * tdist(df=nu_true).rvs(n)

# ---------------------------
# 2) log-likelihood function
# ---------------------------
def log_likelihood(mu, sigma, nu, z):
    # sigma>0, nu>0
    if (sigma <= 0) or (nu < 1):
        return -np.inf
    
    # constant term for StudentT:
    # log( Gamma((nu+1)/2) ) - log( Gamma(nu/2) ) - 0.5*log(nu*pi) - log(sigma)
    c = ( gammaln((nu+1)/2) 
        - gammaln(nu/2)
        - 0.5*math.log(nu*math.pi)
        - math.log(sigma) )
    
    ll = 0.0
    for zi in z:
        tval = (zi - mu)/sigma
        ll += c - ( (nu+1)/2 )*math.log( 1 + (tval**2)/nu )
    return ll

# ---------------------------
# 3) log prior
# ---------------------------
def log_prior_mu(mu):
    # Normal(0, 10)
    return -0.5*(mu/10)**2 - math.log(10*math.sqrt(2*math.pi))

def log_prior_sigma(sigma):
    # HalfNormal(1): p(sigma) ~ exp(-sigma^2/2), for sigma>0
    if sigma <= 0:
        return -np.inf
    # ignoring normalization constant doesn't hurt for MCMC, but let's be explicit:
    # half normal has density = (sqrt(2)/sqrt(pi)) * exp(-sigma^2/2)
    return -0.5*(sigma**2) # ignoring the constant factor doesn't affect acceptance ratio

def log_prior_nu(nu, nu_max=100):
    # discrete uniform over 1..nu_max
    if nu<1 or nu>nu_max:
        return -np.inf
    # each integer equally likely => log(1/nu_max)
    return -math.log(nu_max)

def log_post(mu, sigma, nu, z, nu_max=100):
    return (
       log_likelihood(mu, sigma, nu, z)
     + log_prior_mu(mu)
     + log_prior_sigma(sigma)
     + log_prior_nu(nu, nu_max)
    )

# ---------------------------
# 4) Metropolis-within-Gibbs
# ---------------------------
n_iter = 5000
burn_in = 1000
nu_max = 100

chain_mu = np.zeros(n_iter)
chain_sigma = np.zeros(n_iter)
chain_nu = np.zeros(n_iter, dtype=int)

# initial guesses
cur_mu = 0.0
cur_sigma = 1.0
cur_nu = 5

cur_log_post = log_post(cur_mu, cur_sigma, cur_nu, z_data, nu_max)

# step sizes
proposal_sd_mu = 0.3
proposal_sd_sigma = 0.2

accept_mu = 0
accept_sigma = 0
accept_nu = 0

for i in range(n_iter):
    # -- A) update mu --
    prop_mu = np.random.normal(cur_mu, proposal_sd_mu)
    prop_lp = log_post(prop_mu, cur_sigma, cur_nu, z_data, nu_max)
    log_alpha = prop_lp - cur_log_post
    if (log_alpha >= 0) or (np.log(np.random.rand()) < log_alpha):
        cur_mu = prop_mu
        cur_log_post = prop_lp
        accept_mu += 1
    
    # -- B) update sigma --
    # naive approach: normal proposal, might propose negative
    # could do log-scale to ensure positivity
    prop_sigma = np.random.normal(cur_sigma, proposal_sd_sigma)
    prop_lp = log_post(cur_mu, prop_sigma, cur_nu, z_data, nu_max)
    log_alpha = prop_lp - cur_log_post
    if (log_alpha >= 0) or (np.log(np.random.rand()) < log_alpha):
        cur_sigma = prop_sigma
        cur_log_post = prop_lp
        accept_sigma += 1

    # -- C) update nu (discrete) --
    # random-walk approach: nu +/- 1 or stay
    step_nu = np.random.choice([-1, 0, +1])
    prop_nu = cur_nu + step_nu
    prop_lp = log_post(cur_mu, cur_sigma, prop_nu, z_data, nu_max)
    log_alpha = prop_lp - cur_log_post
    if (log_alpha >= 0) or (np.log(np.random.rand()) < log_alpha):
        cur_nu = prop_nu
        cur_log_post = prop_lp
        accept_nu += 1

    # record
    chain_mu[i] = cur_mu
    chain_sigma[i] = cur_sigma
    chain_nu[i] = cur_nu

# acceptance rates
print(f"Acceptance mu: {accept_mu/n_iter:.2f}")
print(f"Acceptance sigma: {accept_sigma/n_iter:.2f}")
print(f"Acceptance nu: {accept_nu/n_iter:.2f}")

# discard burn-in
mu_samples = chain_mu[burn_in:]
sigma_samples = chain_sigma[burn_in:]
nu_samples = chain_nu[burn_in:]

# ---------------------------
# 5) Posterior Summaries
# ---------------------------
print("Posterior mean(mu) =", np.mean(mu_samples))
print("Posterior mean(sigma) =", np.mean(sigma_samples))
# for nu, maybe a mode or median:
from collections import Counter
counts = Counter(nu_samples)
nu_mode = counts.most_common(1)[0][0]
print("Posterior mode(nu) =", nu_mode)

# ---------------------------
# 6) Posterior Predictive
# ---------------------------
# Let's generate "replicated" data from the posterior draws:
# We'll draw 1 replicate per iteration, for simplicity
replicates = []
for i in range(len(mu_samples)):
    # pick the i-th posterior sample
    m = mu_samples[i]
    s = sigma_samples[i]
    nu_val = nu_samples[i]
    # simulate n data from t_{nu_val}
    # location= m, scale= s
    rep = m + s * tdist(df=nu_val).rvs(size=1)
    replicates.append(rep[0])
replicates = np.array(replicates)

plt.figure(figsize=(10,5))

# Original data histogram vs. replicated
plt.subplot(1,2,1)
plt.hist(z_data, bins=30, density=True, alpha=0.7, label="Observed z_data")
plt.title("Observed Data (n=100)")

plt.subplot(1,2,2)
plt.hist(replicates, bins=30, density=True, alpha=0.7, color='orange', label="Replicates (1 per draw)")
plt.title("Posterior Predictive (Single Rep per Iter)")

plt.tight_layout()
plt.show()
```
